```python
group_id = 0
data = []
group_rows = []
with open("../dataset/HI-Small_Patterns.txt", "r") as f:
    row = []
    for line in f:
        if line == "\n":
            continue

        if line.startswith("BEGIN"):
            group_id += 1
            continue

        if line.startswith("END"):
            method = line.split(" - ")[1]
            for r in group_rows:
                r.append(group_id)
                r.append(method.strip())
            data.extend(group_rows)
            group_rows = []
            continue

        row = line.split(',')
        group_rows.append(row)

        
df = pd.DataFrame(data, columns=['Timestamp', 'From Bank', 'From Account', 'To Bank', 'To Account',
       'Amount Received', 'Receiving Currency', 'Amount Paid',
       'Payment Currency', 'Payment Format', 'Is Laundering', 'Group ID', 'method'])
df.head()    
```
