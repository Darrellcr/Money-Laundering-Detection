```python
import math
from pathlib import Path

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset_path = Path("../dataset")
accounts_path = dataset_path / "HI-Small_accounts.csv"
sample_transactions_path = dataset_path / "transactions_test_sample.csv"
all_transactions_path = dataset_path / "HI-Small_Trans.csv"
print(f"{accounts_path=}")
print(f"{sample_transactions_path=}")
print(f"{all_transactions_path=}")
```

```python
sample_transactions = pd.read_csv(sample_transactions_path, header=None)
all_transactions = pd.read_csv(all_transactions_path)
accounts = pd.read_csv(accounts_path)
```

```python
sample_transactions.head(20)
```

```python
df1 = sample_transactions[[1, 2]].drop_duplicates()
df2 = sample_transactions[[3, 4]].set_axis([1, 2], axis=1).drop_duplicates()
sample_accounts= pd.concat([df1, df2], ignore_index=True).drop_duplicates()
sample_accounts = sample_accounts.set_axis(["Bank ID", "Account Number"], axis=1)
sample_accounts.head(20)
```

```python
sample_accounts = accounts.merge(sample_accounts, on=["Bank ID", "Account Number"], how="inner")
sample_accounts.head(20)
```

```python
sample_accounts.to_csv("../dataset/accounts_test_sample.csv", index=False)
```
